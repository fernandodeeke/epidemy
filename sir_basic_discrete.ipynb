```python
# Basic SIR Discrete Model
import numpy as np
import matplotlib.pyplot as plt

# Model parameters
beta = 1.5          # Infectiousness (transmission rate)
gamma = 0.12        # Recovery rate
T = 50              # Simulation length (days)
S0 = 4999           # Initial number of susceptibles
I0 = 1              # Initial number of infected
R0 = 0.0            # Initial number of recovered

# Initialize state variables
steps = T + 1  # Includes the initial point
S = np.zeros(steps)
I = np.zeros(steps)
R = np.zeros(steps)
time = np.linspace(0, T, steps)  # Time vector

# Initial state
S[0] = S0
I[0] = I0
R[0] = R0
N = S0 + I0 + R0

# Iteration
for t in range(steps - 1):
    S[t + 1] = S[t] - beta * S[t] * I[t] / N
    I[t + 1] = I[t] + (beta * S[t] * I[t] / N - gamma * I[t])
    R[t + 1] = R[t] + gamma * I[t]

# Plot
plt.figure(figsize=(8, 5), dpi=120)
plt.plot(time, S / N, label="S (susceptible)", lw=1.8)
plt.plot(time, I / N, label="I (infected)", lw=1.8)
plt.plot(time, R / N, label="R (recovered)", lw=1.8)
plt.title("Basic discrete SIR model")
plt.xlabel("t(days)")
plt.ylabel("Proportion of the population")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig("sir_basic_discrete.png", dpi=300)
plt.show()
```
